```python
import glob
import os
import matplotlib.pyplot as plt
from osgeo import gdal
from osgeo import ogr
from osgeo import osr
import datetime as dt
import numpy as np

from matplotlib.colors import ListedColormap
colors = ['blue','lime','red']
cmap = ListedColormap(colors)
```

```python
# open up the probability layers for 2019
opn1 = gdal.Open('landcover_full_tile_classification_2019.tif')
arr1 = opn1.ReadAsArray()

# open up the probability layers for 2022
opn2 = gdal.Open('landcover_full_tile_classification_2022.tif')
arr2 = opn2.ReadAsArray()
```

```python
fig,axs = plt.subplots(1,3,figsize=(20,6))
axs[0].imshow(arr1[1])
axs[1].imshow(arr2[1])
im = axs[2].imshow(arr2[1] - arr1[1],vmin=-0.5,vmax=0.5)
plt.colorbar(im)

axs[0].set_title('Forest Porbability 2019')
axs[1].set_title('Forest Porbability 2022')
axs[2].set_title('Proability difference 2022 - 2019')

# when we difference the probability layers, regions that have a positive 
# value have gone up in proability. In this caser it means the probability 
# of forest landcover has gone up. Negative values mean a drop in probability
# which probability means deforestion
```
